# Logistics Regression : Classfication problem

# 1. import necessary libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# 2.import data

```python
gender_classfication = pd.read_csv('weight-height.csv')
```

```python
gender_classfication
```

# 3. Data Understanding

```python
gender_classfication.shape
```

```python
gender_classfication.isnull().sum()
```

# 4.Data Preparation

```python
gender_classfication.head()
```

```python
gender_classfication['Gender'] = gender_classfication['Gender'].replace(to_replace=['Male','Female'],value=[0,1])
```

```python
gender_classfication
```

```python
gender_classfication.dtypes
```

```python
gender_classfication['Height'] = gender_classfication['Height']*2.54
gender_classfication['Weight'] = gender_classfication['Weight']/2.205
```

# 5. Model Building

```python
gender_classfication.drop(['Gender'],axis=1)
```

```python
gender_classfication
```

```python
x = gender_classfication.drop(['Gender'],axis = 1)
y = gender_classfication[['Gender']]
```

# 6. Model Training

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x,y)
```

```python
logistic_model.coef_
```

```python
logistic_model.intercept_
```

# 7. Model Testing

```python
y_pred = logistic_model.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

# 8. Model Deployment

```python
from pickle import dump
```

```python
dump(logistic_model,open('gender_classfication_intelligence.pkl','wb'))
```
