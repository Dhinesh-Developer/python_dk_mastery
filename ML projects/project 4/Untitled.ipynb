# Titanic Survival Prediction using NAIVE BAYES

# 1. Importing libraries

```python
import pandas as pd
import numpy as np
```

# 2. load the dataset

```python
dataset = pd.read_csv('titanicsurvival.csv')
```

```python
dataset
```

# 3. Summarize dataset

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.head(5)
```

# 4. Mapping Text data to binary Value

```python
income_set = set(dataset['Gender'])
dataset['Gender'] = dataset['Gender'].map({'female':0,'male':1}).astype(int)
print(dataset.head())
```

```python
dataset.dtypes
```

# 5. Segregate Dataset into X(input) and Y(output)

```python
X = dataset.drop('Survived',axis='columns')
X
```

```python
Y = dataset['Survived']
Y
```

# 6. Finding and removing NA values from our Features X

```python
X.columns[X.isna().any()]
```

```python
X.Age = X.Age.fillna(X.Age.mean())
```

```python
dataset
```

# 7. Test again to check any na value

```python
X.columns[X.isna().any()]
```

# 8. Splitting Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state = 0)
```

# 9. Training 

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

# 10. Prediction for all test Data

```python
y_pred = model.predict(X_test)
print(np.column_stack((y_pred,y_test)))
```

# 11. Accuracy of the Model

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_pred)*100))
```

# 12. Predicting,wheather person Survived or not 

```python
pclassNo = int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Person's Gender 0-female/ 1-male: "))
age = int(input("Enter Person's Age: "))
fare = float(input("Enter Person's Fare: "))
person = [[pclassNo,gender,age,fare]]
res = model.predict(person)
print(res)

if res == 1:
    print("Person might be Survived!")
else:
    print("Person might not be Survived!")
```
