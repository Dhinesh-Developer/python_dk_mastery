##  4.Segerate dataset into X(input) & Y(output)

```python
# before comma rows and after comma columns

# X = dataset.iloc[rowstart:rowend,colstart:colend]
# above form the below data is give. iloc -> means index location
X = dataset.iloc[:, :-1].values
X
```

```python
Y = dataset.iloc[:, -1].values
Y
```

## 5. Splitting Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
# 0.25 na testing 25% and training 80% ,, random_state => meaning random 75% data from the dataset
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state = 0)
```

### Feature Scaling 
- we scale our data to make all the features contribute equally to the result
- Fit_transform - fit method is calculating the mean and variance of each of the features persent in our data
- Transform - Transform method is transforming all the features using the respective mean and variance
- we want our test data to be a completely new and surprise set of our model

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # always give fit for training
X_test = sc.transform(X_test)
```

```python
X_train
```

## 6. Model Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train,y_train)
```

### Predicting whether new customer with Age & Salary will buy or not

```python
age = int(input("Enter New Customer Age: "))
sal = int(input("Enter New Customer Salary: "))
newCust = [[age,sal]]
res = model.predict(sc.transform(newCust))
print(res)
if res == 1:
    print("customer will buy!")
else:
    print("customer won't buy!")
```

## Prediction for all Test data

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Accuracy score

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_pred)*100))
```
