# Linear Regression, Ridge and Lasso

```python
# hpuse pricing dataset
```

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
type(data)
```

```python
data
```

```python
dataset = pd.DataFrame(data.data)
dataset.columns = data.feature_names
dataset.head(5)
```

```python
dataset['price'] = data.target
```

```python
dataset.head()
```

```python
## dividing the dataset into independent and dependent features
X = dataset.iloc[:, :-1] ## independent features
y = dataset.iloc[:,-1]## dependent features
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state=0)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
#mean square error
mse = cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse = np.mean(mse)
```

```python
print(mean_mse)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

params = {'alpha': [0.001, 0.01, 0.1, 1, 5, 10, 20, 50, 100]}

ridge_regressor = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# 🔹 Step 1: Scale the features (VERY IMPORTANT)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 🔹 Step 2: Create Lasso model with higher iterations
lasso = Lasso(max_iter=10000)

# 🔹 Step 3: Use proper alpha values (avoid very small numbers)
params = {
    'alpha': [0.001, 0.01, 0.1, 1, 5, 10, 20, 50, 100]
}

# 🔹 Step 4: GridSearchCV
lasso_regressor = GridSearchCV(
    lasso,
    params,
    scoring='neg_mean_squared_error',
    cv=5
)

# 🔹 Step 5: Fit model
lasso_regressor.fit(X_train, y_train)

# 🔹 Step 6: Results
print("Best Alpha:", lasso_regressor.best_params_)
print("Best Score:", lasso_regressor.best_score_)
```

# Logisitice Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
# Independent features
X = pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
## Dependent features
y = pd.DataFrame(df['target'],columns=['Target'])
y
```

```python
y['Target'].value_counts()
```

```python
## Train Test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state=0)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=100,max_iter=100)
```

```python
model = GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
