```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = """Hello Welcome,to dhinesh kumar's NLP Tutorials.
Please do watch the entire coursel to become expert in NLP.
"""
```

```python
corpus
```

```python
## Tokenization
## Sentence --> paragraphs

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization
## Paragraph --> words
## Sentence --> words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```
