# wordnet Lemmatizer 

```python
from nltk.stem import WordNetLemmatizer
# Download required resources
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going",pos='v')
```

```python
"""
POS Noun - n
verb - v
adjective - a
adverb - r
"""
```

```python
words = ["eating","eats","eaten","writing","writes","programming","programs","going","goes","final","history","finally"]
```

```python
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("goes")
```

```python
lemmatizer.lemmatize("failry",pos='v')
```
