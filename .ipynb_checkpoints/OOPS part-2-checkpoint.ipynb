## Create class and object
- class → blueprint
- __init__ → constructor
- self → current object

```python
class User:
    def __init__(self,name):
        self.name = name

    def show(self):
        print("Name: ",self.name)

# creating object
u1 = User("Kumar")
u1.show()
```

### Encapsulation

```python
class User:
    def __init__(self,pin):
        self.__pin = pin # double underscore na private, single underscore na protected

    def chech_pin(self,input_pin):
        if input_pin == self.__pin:
            print("Correct PIN!")
        else:
            print("OOPS! Wrong PIN!")

u = User(7777)
u.chech_pin(7777)
```

### INHERITANCE

```python
class Person:
    def __init__(self,name):
        self.name = name

    def show_name(self):
        print("Name:",self.name)

class Student(Person):
    def study(self):
        print("Studying...")

s = Student("Kumar")
s.show_name()
s.study()
```

### POLYMORPHISM

```python
class Dog:
    def sound(self):
        print("Bark..")

class Cat:
    def sound(self):
        print("Meow..")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

### ABSTRACTION

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started")

v = Car()
v.start()
    
```

## SIMPLE COMBINATION

```python
from abc import ABC, abstractmethod

class ATM(ABC):

    def __init__(self, pin):
        self.__pin = pin   # encapsulation

    def check_pin(self, p):
        return p == self.__pin

    @abstractmethod
    def transaction(self):
        pass


class SBI(ATM):   # inheritance

    def transaction(self):   # polymorphism
        print("SBI Transaction")


user = SBI(1234)

print(user.check_pin(1234))
user.transaction()
```
