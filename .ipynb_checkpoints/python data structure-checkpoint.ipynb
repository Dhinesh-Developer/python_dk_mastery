- indexing ans slicing is possible

```python
student_mark
```

```python
type (student_mark)
```

```python
empty_cart = []
```

```python
empty_cart
```

```python
items_to_buy = ["Apple","Orange","Graphes","Gauva","Apple"]
items_to_buy
```

```python
items_to_buy = ["Apple","Orange","Graphes","Gauva","Apple",6.75,10]
items_to_buy
```

```python
# slicing is done by index start from 0, using the len function return the number of items in a contains
```

```python
len(items_to_buy)
```

```python
items_to_buy[0] # indexing
```

```python
items_to_buy[0:2] # start = 0, end = len+1 always use -> data is start to end index -1 , this concept called as range of valuse
```

```python
items_to_buy[-3] # negative indexing is also positive -> traverse from right to left
```

```python
items_to_buy[2:6]
```

```python
items_to_buy[-7:-4] # end : start
```

```python
#changing the values in future
items_to_buy[1] = "Watermelon" # this process is called mutable
```

```python
items_to_buy
```

## Sets
- set does not allow duplicates
- 1D
- it is hetergenous
- indexing and Slicing is not possible
- set is mutable

```python
secret_values = 723,234,234.234
```

```python
secret_values
```

```python
empty_cart2 = set()
```

```python
empty_cart2
```

```python
type(empty_cart2)
```

```python
items_to_buy2 = {"Apple","Orange","Graphes","Gauva","Apple",23}
items_to_buy2 # set does not allow duplicates && it is one-dimensional && it is hetergenous 
```

```python
items_to_buy2[2] # 'set' object is not subscriptable -> means that indexing and Slicing is not possible
```

```python
# how set is mutable?
# adding an element
items_to_buy2.add("banana")

#removing an element
items_to_buy2.discard("Graphes")
```

```python
items_to_buy2
```

```python
items_to_buy2.append("Red banana")
```
