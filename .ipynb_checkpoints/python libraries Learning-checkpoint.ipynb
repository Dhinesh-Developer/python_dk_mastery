## 2. Numpy - Numerical Operations

```python
!pip install numpy
```

```python
import numpy as np
```

```python
# repeat is used the repeat the data in row or column how many number of times
np.repeat(a = medical_data, repeats = 2,axis=0)
# axis = 0 na row,1 na column
# this is array form
```

```python
# this is table form-> using the DataFrame
medical_table_data = pd.DataFrame(data = np.repeat(a = medical_data, repeats = 2,axis=0))
```

```python
medical_table_data
```

## Types of datatypes:
- Discrete - Categories, just countable, it will never have any units, you can't measure that
- Continous - It will have units, it is measure able

```python
# hist-> histogram only allows continous data
# units -> means number, discrete means categorires

# now calucaluating the age
plt.hist(x = medical_data.Age)
```
