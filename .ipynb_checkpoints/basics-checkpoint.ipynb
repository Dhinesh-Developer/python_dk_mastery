## print if the user guessed number is right or not

```python
secret_number = 56
# type casting -> converting one datatype to another data types.
user_input = int(input("Enter your guess: "))
if secret_number == user_input:
    print("congralutaions!")
else:
    print("Guess is wrong")
```

## level 2 : print if the user guessed number is right or not but with max of 7 trails

#### 2.1 using for loop

```python
secret_number = 56
# type casting -> converting one datatype to another data types.

# always give end + 1
for i in range(0,7):
    user_input = int(input("Enter your guess: "))
    if secret_number == user_input:
        print("congralutaions!")
        break
    else:
        print("Guess is wrong")
```

#### 2.2 using while loop

```python
secret_number = 56
# type casting -> converting one datatype to another data types.

# always give end + 1
while True:
    user_input = int(input("Enter your guess: "))
    if secret_number == user_input:
        print("congralutaions!")
        break
    else:
        print("Guess is wrong")
```

### level 3: Conditionals statement and functions

1.User should enter the number between 1-100 .If it's entered wrong,the program shoul say, "Invalid number"
2. If the user guessed number> secret number, then program should say, "You number is greater than the secret number"
3. If the user guessed number< secret number, then program should say, "You number is lesser than the secret number"

```python
print("Welcome to Number Guessing Game.!!")
print("Guess the Number between 1 to 100")

secret_number = 98

for i in range(1,8):
    user_input = int(input("Enter the Number: "))
    if user_input < 1 or secret_number > 100:
        print("Invalid number! please enter the number between 1-100")
    elif user_input > secret_number:
        print("Your number is higher")
    elif user_input < secret_number:
        print("Your number is lesser")
    else:
        print("Guessed Number is Correct!!!!")
        print(f"Congralutaions, You have guessed the number {secret_number} in {i} attempts")
        break
        
    
```
