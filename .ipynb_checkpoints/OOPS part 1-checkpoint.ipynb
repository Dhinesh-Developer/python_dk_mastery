# Real World analogy -1
 #### i want to reuse the same function for Team Members and Managers,but they have small differences. How do i do that?
 - Both Team Members and Managers should be able to log in,but
 - Team Members -> should log in, view tasks, and submit tasks.
 - Managers -> Should log in, view tasks, and assign tasks.

```python
### WITHOUT OOPS

#Defininig a function
def login(user_data):
    print(f"{user_data['name']} has logged in!")

def view_tasks(user_data):
    print(f"{user_data['name']} is viewing the task")

def submit_tasks(user_data):
    if user_data["role"] == "team_members":
        print(f"{user_data['name']} has submitted the task")
    else:
        print("manager no need to submit the task")

def assign_tasks(user_data):
    if user_data["role"] == "manager":
        print(f"{user_data['name']} has assigned the task")
    else:
        print("Team members no need to assign the task")    

# DATA COLLECTION

# user_data = {"name":"kumar","role":"manager"}
user_data = {"name":"dharshini","role":"team_member"}
# function to called
login(user_data)
view_tasks(user_data)
submit_tasks(user_data)
assign_tasks(user_data)
```

### Note: Importance of Self and Init comes into the picture
- Encapsulation : _ single underscore na protected and __ double underscore na private and then no underscore na public

```python
# creating a class
class SecureUser:
    def __init__(self,name,atm_pin):
        self.name = name
        self._atm_pin = atm_pin
        
    def get_pin(self,attempt):
        return attempt == atm_pin

    def deposit_cash(self):
        pass

    def withdraw_cash(self):
        pass

# creating a object
kumar_data = SecureUser("kumar",7777)
# user_data.get_pin(user_data,2131)
```
