### 2. Create a list of numbers and print only even numbers using a loop

```python
numbers = range(1,20)
even_numbers = []
for i in numbers:
    if i%2 == 0:
        even_numbers.append(i)
even_numbers        
```

### 3. Write a program that handles division by zero using try-except

```python
try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))
    res = numerator/denominator
    print(res)
except ZeroDivisionError:
    print("Invalid Entry! Enter a number orther than zero!")
except ValueError:
    print("Invalud Entry! Enter a number not alphabets!")
finally:
    print("Program Ended!") 
    
```

### 4.Given a String, count how many vowels it has.

```python
user_input = input("Enter any word or sentence: ")
cnt = 0
for i in user_input.lower():
    if i in 'aeiou':
        cnt = cnt+1
cnt        
```

### 5.Reverse a list without using built-in functions

```python
original_list = [1,2,3,4,5]
new_list = []
l=0
r = len(original_list)-1
while l<=r:
    temp = original_list[l]
    original_list[l] = original_list[r]
    original_list[r] = temp;
    l = l+1;
    r = r-1;
original_list    
    
```

### Replace all missing values in a DataFrame with the column's mean
