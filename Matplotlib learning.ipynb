```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("medical_data_100.csv")
```

```python
df
```

```python
plt.plot(df["Age"])
plt.show()
```

```python
plt.plot(df["Diagnosis"],df["Blood_Group"])
plt.show()
```

```python
plt.plot(df["Diagnosis"],df["Blood_Group"])
plt.title("Graph Analysis of Blood group with diagnosis")
plt.ylabel("Blood_Group")
plt.xlabel("Diagnosis")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot("Patient_Name")
plt.show()
```

```python
# Histogram graph

plt.hist(df["Diagnosis"])
plt.title("Analysis of Diagnosis")
plt.xlabel("Doctor")
# plt.ylabel("Diagnosis")
plt.show()
```

```python
# grid graph

plt.hist(df["Diagnosis"])
plt.grid()
plt.title("Analysis of Diagnosis")
plt.xlabel("Doctor")
# plt.ylabel("Diagnosis")
plt.show()
```

```python
# normal plot with legend
plt.plot(df["Age"],label="Age")
plt.plot(df["Doctor"],label='Doctor')
plt.legend()
plt.show()
```

```python
# changing the color, linestyle and then marker
plt.plot(df["Diagnosis"],label ="Diagnosis",linestyle='--',marker='o')
plt.plot()
```

```python
plt.plot(df["Age"])
plt.savefig("graph.png")
```

```python
# bar graph
disease_count = df["Diagnosis"].value_counts()
plt.bar(disease_count.index,disease_count.values)
plt.show()
```

```python
disease_count = df["Age"].value_counts()
plt.pie(disease_count.values,labels = disease_count.index,autopct="%1.1f%%")
plt.title("Age Analysis")
plt.show()
```

```python
disease_count = df["Diagnosis"].value_counts()
plt.pie(disease_count.values,labels=disease_count.index,autopct="%1.2f%%")
plt.title("Diagnosis Analysis")
plt.show()
```
